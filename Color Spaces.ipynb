```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pic  = cv2.imread("G:/c1.jpg")
cv2.imshow("image", pic)
cv2.waitKey(0)
```

```python
gray = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)
cv2.imshow("gray", gray)
cv2.waitKey(0)
```

```python
hsv = cv2.cvtColor(pic, cv2.COLOR_BGR2HSV)
cv2.imshow("hsv", hsv)
cv2.waitKey(0)
```

```python
lab = cv2.cvtColor(pic, cv2.COLOR_BGR2LAB)
cv2.imshow("lab", lab)
cv2.waitKey(0)
```

```python
plt.imshow(lab)
plt.show()
```

```python
hsv_bgr = cv2.cvtColor(pic, cv2.COLOR_HSV2BGR)
plt.imshow(hsv_bgr)
plt.show()
```

```python
lab_bgr = cv2.cvtColor(pic, cv2.COLOR_LAB2BGR)
plt.imshow(lab_bgr)
plt.show()
```
