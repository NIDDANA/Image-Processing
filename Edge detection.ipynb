```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pic  = cv2.imread("G:/c1.jpg")
cv2.imshow("image", pic)
cv2.waitKey(0)
```

```python
gray = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)
cv2.imshow("gray", gray)
cv2.imshow("image", pic)

cv2.waitKey(0)
```

```python
# laplation

lap = cv2.Laplacian(gray, cv2.CV_64F)
lap = np.uint8(np.absolute(lap))
cv2.imshow("laplation", lap)
cv2.imshow("image", pic)
cv2.imshow("gray", gray)

cv2.waitKey(0)

```

```python
#sabel 
sabelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0) 
sabely = cv2.Sobel(gray, cv2.CV_64F, 0, 1)

cv2.imshow("sobalx", sabelx)
cv2.imshow("sabely", sabely)

cv2.waitKey(0)
```

```python
plt.imshow(sabelx)
plt.imshow(sabely)
plt.show()
```
